```python
import seaborn as sns
import math
```

```python
res = []
```

```python
counter = 0

def z2(a, counter):
    if counter > 1000:
        return
    counter += 1
    a = a**2
    res.append(a)
    return z2(a, counter)
    
```

```python
x = [x.real for x in res if math.isnan(x.real) == False ]
y = [y.imag for y in res if math.isnan(y.imag) == False]
```

```python
sns.lineplot(x=x, y=y, markers="o", label="complex number")
```
